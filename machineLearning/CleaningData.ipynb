###3. Bin Data

```python
#make column with binned areas for sDA
sDABin = pd.qcut(dataMod['sDA'], q=5, duplicates='drop')
dataMod['sDABin'] = sDABin

#make column with binned areas for UDI
UDIBin = pd.qcut(dataMod['UDI'], q=5)
dataMod['UDIBin'] = UDIBin

#make column with binned areas for ASE
ASEBin = pd.qcut(dataMod['ASE'], q=5, duplicates='drop')
dataMod['ASEBin'] = ASEBin

dataMod
```

```python
#copy these results in to the next cell
np.unique(dataMod['sDABin'])
```

```python
#change bin interval to categroical name for sDA
binSize_sDA = []
for value in dataMod['sDABin']:
  if value == pd.Interval(-0.001, 12.3, closed='right'):
    binSize_sDA.append('1')
  elif value == pd.Interval(12.3, 38.67, closed='right'):
    binSize_sDA.append('2')  
  elif value == pd.Interval(38.67, 80.77, closed='right'):
    binSize_sDA.append('3')  
  else:
    binSize_sDA.append('4')  
```

```python
#copy these results in to the next cell
np.unique(dataMod['UDIBin'])
```

```python
#change bin interval to categroical name for UDI
binSize_UDI = []
for value in dataMod['UDIBin']:
  if value == pd.Interval(-0.001, 29.653, closed='right'):
    binSize_UDI.append('1')
  elif value == pd.Interval(29.653, 53.629, closed='right'):
    binSize_UDI.append('2')  
  elif value == pd.Interval(53.629, 73.968, closed='right'):
    binSize_UDI.append('3')  
  elif value == pd.Interval(73.968, 86.117, closed='right'):
    binSize_UDI.append('4')  
  else:
    binSize_UDI.append('5')  
```

```python
#copy these results in to the next cell
np.unique(dataMod['ASEBin'])
```

```python
#change bin interval to categroical name for ASE
binSize_ASE = []
for value in dataMod['ASEBin']:
  if value == pd.Interval(-0.001, 0.858, closed='right'):
    binSize_ASE.append('1')
  elif value == pd.Interval(0.858, 12.32, closed='right'):
    binSize_ASE.append('2')  
  elif value == pd.Interval(12.32, 38.71, closed='right'):
    binSize_ASE.append('3')  
  else:
    binSize_ASE.append('4')  
```

```python
dataMod['sDA_Category'] = binSize_sDA
dataMod['UDI_Category'] = binSize_UDI
dataMod['ASE_Category'] = binSize_ASE

dataMod
```
